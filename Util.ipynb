```python
import matplotlib.pyplot as plt
import numpy as np
sigmoid = lambda x: 2*(1 / (1 + np.exp(-2*x)))-1

y=np.linspace(-10,10,100)

plt.plot(y,sigmoid(y))
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('ReLU')

plt.text(3,0.8,r'$\sigma(x)=\frac{2}{1+e^{-2x}}-1$',fontsize=15)
plt.legend(loc='lower right')
plt.legend(bbox_to_anchor=(0.5, -0.2), loc='center', borderaxespad=0)

plt.subplots_adjust(bottom=0.22)
```

```python

import matplotlib.pyplot as plt

# Rectified Linear Unit (ReLU)
def ReLU(x):
  data = [max(0.05*value,value) for value in x]
  return np.array(data, dtype=float)



# Generating data for Graph
x_data = np.linspace(-10,10,100)
y_data = ReLU(x_data)


# Graph
plt.plot(x_data, y_data, 'b')
plt.title('ReLU ')
plt.legend(['ReLU','der_ReLU'])
plt.grid()
plt.show()
```
